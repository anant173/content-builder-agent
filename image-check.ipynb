```python
from dotenv import load_dotenv
load_dotenv()
from openai import OpenAI
import base64
import os
```

```python
tfy_base_url = "https://gateway.truefoundry.ai"
tfy_api_key = os.environ.get("TFY_API_KEY")
```

```python
client = OpenAI(
    api_key = tfy_api_key,
    base_url=tfy_base_url
)
```

```python
prompt = "Make a dog"
    
response = client.images.generate(
    model="gcp-vertex-default/gemini-2.5-flash-image",
    prompt=prompt,
    n=1,
    size="1024x1024"
)
```

```python
for part in response.data:
    if hasattr(part, 'url') and part.url:
        print(f"Image URL: {part.url}")
    if hasattr(part, 'b64_json') and part.b64_json:
        image_bytes = base64.b64decode(part.b64_json)
        
        output_path = "hero.png"

        with open(output_path, "wb") as f:
            f.write(image_bytes)
            print(f"Image saved to {output_path}")
```
